```python
# Full corrected pipeline: EDA + Preprocessing + MLP classification (Pass/Fail) on student-mat.csv
# Requirements: pandas, numpy, matplotlib, seaborn, scikit-learn, tensorflow
# Install if needed: pip install pandas numpy matplotlib seaborn scikit-learn tensorflow

import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, confusion_matrix, classification_report,
    roc_curve, auc
)

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# ---------------------------
# Reproducibility (best-effort)
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ---------------------------
# 1. Load data
fn = 'student-mat.csv'   # ensure this file is in working directory
if not os.path.exists(fn):
    raise FileNotFoundError(f"Place '{fn}' in the working directory and re-run.")

# The UCI student dataset uses semicolon-separated values
df = pd.read_csv(fn, sep=';')
print("Loaded data shape:", df.shape)
print(df.head().to_string(index=False))

# ---------------------------
# 2. Quick data understanding
print("\n--- Info ---")
df.info()

print("\n--- Describe (numeric) ---")
print(df.describe().T)

print("\nMissing values per column:")
print(df.isnull().sum())

print("\nDuplicate rows:", df.duplicated().sum())

# ---------------------------
# 3. Create binary target: pass_fail (1 = pass if G3 >= 10, else 0)
df['pass_fail'] = df['G3'].apply(lambda x: 1 if x >= 10 else 0)
print("\nValue counts for pass_fail:")
print(df['pass_fail'].value_counts())

# ---------------------------
# 4. EDA Visualizations (multiple plots)
sns.set(style="whitegrid")
plt.rcParams.update({'figure.max_open_warning': 0})

# Visual 1: Histogram of final grade (G3)
plt.figure(figsize=(6,4))
plt.hist(df['G3'], bins=11)
plt.title('Distribution of Final Grades (G3)')
plt.xlabel('G3')
plt.ylabel('Frequency')
plt.show()

# Visual 2: Histogram side-by-side for G1, G2, G3
plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.hist(df['G1'], bins=10); plt.title('G1 distribution'); plt.xlabel('G1')
plt.subplot(1,3,2); plt.hist(df['G2'], bins=10); plt.title('G2 distribution'); plt.xlabel('G2')
plt.subplot(1,3,3); plt.hist(df['G3'], bins=10); plt.title('G3 distribution'); plt.xlabel('G3')
plt.tight_layout(); plt.show()

# Visual 3: Scatter G1 vs G3
plt.figure(figsize=(6,4))
plt.scatter(df['G1'], df['G3'], alpha=0.6)
plt.title('G1 vs G3')
plt.xlabel('G1')
plt.ylabel('G3')
plt.show()

# Visual 4: Scatter G2 vs G3
plt.figure(figsize=(6,4))
plt.scatter(df['G2'], df['G3'], alpha=0.6)
plt.title('G2 vs G3')
plt.xlabel('G2')
plt.ylabel('G3')
plt.show()

# Visual 5: Boxplot - Gender vs G3
plt.figure(figsize=(6,4))
sns.boxplot(x='sex', y='G3', data=df)
plt.title('Gender vs Final Grade (G3)')
plt.show()

# Visual 6: Study time vs average G3
plt.figure(figsize=(6,4))
sns.barplot(x='studytime', y='G3', data=df, ci=None)
plt.title('Studytime vs Avg Final Grade')
plt.xlabel('Study time category (1..4)')
plt.show()

# Visual 7: Absences vs G3 (boxplot)
plt.figure(figsize=(8,4))
sns.boxplot(x='absences', y='G3', data=df)
plt.title('Absences vs Final Grade (G3)')
plt.show()

# Visual 8: Countplot failures
plt.figure(figsize=(6,4))
sns.countplot(x='failures', data=df)
plt.title('Count of Past Failures')
plt.show()

# Visual 9: Pairplot for G1, G2, G3 (colored by pass_fail)
try:
    sns.pairplot(df[['G1', 'G2', 'G3', 'pass_fail']], hue='pass_fail', corner=True)
    plt.suptitle('Pairplot: G1, G2, G3 colored by pass_fail', y=1.02)
    plt.show()
except Exception as e:
    print("Pairplot skipped (may be slow in some environments). Error:", e)

# Visual 10: Correlation heatmap
plt.figure(figsize=(14,10))
corr = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=False)
plt.title('Correlation Matrix (numeric features)')
plt.show()

# ---------------------------
# 5. Preprocessing
# 5.1 Prepare features and target
X = df.drop(columns=['G3', 'pass_fail'])  # keep G1 and G2 as predictors
y = df['pass_fail']

# 5.2 Identify categorical columns
cat_cols = X.select_dtypes(include=['object']).columns.tolist()
num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
print("\nCategorical columns:", cat_cols)
print("Numeric columns:", num_cols)

# 5.3 One-hot encode categorical columns
X_encoded = pd.get_dummies(X, columns=cat_cols, drop_first=True)

# ---------------------------
# === CRITICAL FIX: Ensure all columns are numeric and handle any non-numeric artifacts ===
X_encoded = X_encoded.apply(pd.to_numeric, errors='coerce')  # convert any stray objects to numeric (NaN if impossible)
X_encoded = X_encoded.fillna(0)  # fill any NaNs introduced by coercion

print("\nFeature matrix shape after encoding and conversion:", X_encoded.shape)
print("Any remaining object dtypes?:", X_encoded.select_dtypes(include=['object']).columns.tolist())

# 5.4 Train-test split (use stratify to keep pass/fail balance)
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=SEED, stratify=y
)
print("\nTrain shape:", X_train.shape, "Test shape:", X_test.shape)

# 5.5 Scale numeric columns (fit scaler on training only)
scaler = StandardScaler()
# scale all numeric columns (after get_dummies all columns are numeric)
numeric_after_encoding = X_encoded.columns.tolist()

# Fit scaler using DataFrame to preserve column names and avoid SettingWithCopyWarning
scaler.fit(X_train[numeric_after_encoding])
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

X_train_scaled[numeric_after_encoding] = scaler.transform(X_train[numeric_after_encoding])
X_test_scaled[numeric_after_encoding] = scaler.transform(X_test[numeric_after_encoding])

# Convert to numpy arrays for Keras
X_train_np = X_train_scaled.values.astype(np.float32)
X_test_np = X_test_scaled.values.astype(np.float32)
y_train_np = y_train.values.astype(np.int32)
y_test_np = y_test.values.astype(np.int32)

# ---------------------------
# 6. Build MLP model (binary classification)
input_dim = X_train_np.shape[1]
print("\nInput feature dimension:", input_dim)

model = Sequential([
    Dense(128, activation='relu', input_shape=(input_dim,)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')   # sigmoid for binary classification -> outputs prob
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# Early stopping to prevent overfitting
es = EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True)

# ---------------------------
# 7. Train the model
EPOCHS = 60
BATCH = 32

history = model.fit(
    X_train_np, y_train_np,
    validation_split=0.2,
    epochs=EPOCHS,
    batch_size=BATCH,
    callbacks=[es],
    verbose=2
)

# ---------------------------
# 8. Plot training curves
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title('Loss vs Epoch')
plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='train_acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.title('Accuracy vs Epoch')
plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend()
plt.tight_layout()
plt.show()

# ---------------------------
# 9. Predictions & Evaluation
y_prob = model.predict(X_test_np).ravel()   # probability of positive class
y_pred = (y_prob >= 0.5).astype(int)        # threshold 0.5 to get class labels

# Basic metrics
acc = accuracy_score(y_test_np, y_pred)
print(f"\nTest Accuracy: {acc:.4f}")

print("\nConfusion Matrix:")
cm = confusion_matrix(y_test_np, y_pred)
print(cm)

# Confusion matrix plot
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Fail','Pass'], yticklabels=['Fail','Pass'])
plt.ylabel('True'); plt.xlabel('Predicted'); plt.title('Confusion Matrix')
plt.show()

# Classification report
print("\nClassification Report:")
print(classification_report(y_test_np, y_pred, target_names=['Fail','Pass']))

# ROC Curve & AUC (binary)
fpr, tpr, thresholds = roc_curve(y_test_np, y_prob)
roc_auc = auc(fpr, tpr)
print(f"\nAUC: {roc_auc:.4f}")

plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.3f}')
plt.plot([0,1],[0,1],'--', color='gray')
plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate'); plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()

# ---------------------------
# 10. Optional: Permutation importance (best-effort; may be slow)
# We'll attempt a light-weight approach and wrap in try/except to avoid breaking the script.
try:
    from sklearn.inspection import permutation_importance

    # sklearn expects an estimator with a predict method that accepts X as numpy array.
    # We'll build a small wrapper object with predict method that returns probabilities.
    class KerasProbabilityWrapper:
        def __init__(self, keras_model):
            self.model = keras_model
        def predict(self, X):
            # return probability of positive class (shape: (n_samples,))
            return self.model.predict(X).ravel()

    wrapper = KerasProbabilityWrapper(model)
    print("\nComputing permutation importance (this may take a while)...")
    r = permutation_importance(wrapper, X_test_np, y_test_np, n_repeats=10, random_state=SEED, scoring='roc_auc')
    imp_idx = np.argsort(r.importances_mean)[::-1][:15]  # top 15
    top_features = np.array(X_test.columns)[imp_idx]
    plt.figure(figsize=(8,6))
    plt.barh(range(len(imp_idx)), r.importances_mean[imp_idx][::-1], xerr=r.importances_std[imp_idx][::-1])
    plt.yticks(range(len(imp_idx)), top_features[::-1])
    plt.xlabel('Mean decrease in score (AUC) - permutation importance')
    plt.title('Top ~15 Feature Importances (Permutation)')
    plt.show()
except Exception as e:
    print("\nPermutation importance skipped or failed (non-critical). Error:", e)

# ---------------------------
# 11. Save model & scaler if needed (optional)
# model.save('student_passfail_mlp.h5')
# import joblib
# joblib.dump(scaler, 'scaler.save')

# ---------------------------
# 12. Final notes / interpretation (print sample)
print("\nSample predictions (first 10 test instances):")
sample_df = pd.DataFrame({
    'y_true': y_test_np[:10],
    'y_prob': np.round(y_prob[:10], 3),
    'y_pred': y_pred[:10]
})
print(sample_df.to_string(index=False))

print("\nEnd of pipeline. In your report: discuss accuracy, confusion matrix (false positives/negatives), ROC/AUC, class balance, and which features (G1/G2, studytime, absences, failures, etc.) seem influential.")
```
